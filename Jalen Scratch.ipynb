```python
def get_modulus(fname):
    a = numpy.loadtxt(fname,skiprows=1,delimiter=',')
    stop = 10
    x = a[:stop,0]
    y = a[:stop,1]
    
#plt.plot(a[:,0],a[:,1])
    a_1, a_0 = numpy.polyfit(x,y,1)
#print(a_1)
#fit = numpy.poly1d((a_1,a_0))
#plt.plot(a[:stop,0],fit(a[:stop,0]))
    return a_1
moduli = []
for i in range(999):
    fname = "samples/"+str(i)+".raw"
    moduli.append(get_modulus(fname))
```

```python
average = sum(moduli)/len(moduli)#average Young's Modulus
print(average)
```

```python
stdev = numpy.std(moduli)#standard deviation of Young's Modulus
print(stdev)
```
